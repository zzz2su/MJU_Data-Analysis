```python
from scipy import stats
```

```python
s_array = SEOUL.columns.tolist()   # 컬럼명 리스트
s_arr  = []                       

for element in s_array:
    pearsonr, pvalue = stats.pearsonr(SEOUL[element], SEOUL['tot_preference'])
    s_arr.append(pvalue) #r각 컬럼과 선호도 간 pearson test p value 출력

print(s_array)
print(s_arr)
```

```python
s_arr_df = pd.DataFrame(s_arr, columns=['pvalue'])
s_array_df = pd.DataFrame(s_array, columns=['column'])

## 선호도와 컬럼별 pearson 상관계수 분석 결과(pvalue)
s_df = pd.concat([s_array_df, s_arr_df], axis=1)
s_df.sort_values('pvalue', ascending=True)
```

```python
j_array = JEJU.columns.tolist()   # 컬럼명 리스트
j_arr  = []                       

for element in j_array:
    pearsonr, pvalue = stats.pearsonr(JEJU[element], JEJU['tot_preference'])
    j_arr.append(pvalue) #r각 컬럼과 선호도 간 pearson test p value 출력

print(j_array)
print(j_arr)
```

```python
j_arr_df = pd.DataFrame(j_arr, columns=['pvalue'])
j_array_df = pd.DataFrame(j_array, columns=['column'])

## 선호도와 컬럼별 pearson 상관계수 분석 결과(pvalue)
j_df = pd.concat([j_array_df, j_arr_df], axis=1)
j_df.sort_values('pvalue', ascending=True)
```
